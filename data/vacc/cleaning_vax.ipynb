```python
import pandas as pd
import datetime as dt
```

```python
c = pd.read_csv("vax_demog_age.csv")
```

```python
c.tail()
```

```python
c = c.drop(columns=['district'])
```

```python
c = c.groupby(['date','state']).agg("sum")
```

```python
# c = c.drop(columns=["partial_missing",'full_missing','booster_missing','booster2_missing'])
```

```python
c['partial'] = c['partial_5_11'] + c['partial_12_17'] + c['partial_18_29'] + c['partial_30_39'] + c['partial_40_49'] + c['partial_50_59'] + c['partial_60_69'] + c['partial_70_79'] + c['partial_80']+c['partial_missing']
```

```python
c['full'] = c['full_5_11'] + c['full_12_17'] + c['full_18_29'] + c['full_30_39'] + c['full_40_49'] + c['full_50_59'] + c['full_60_69'] + c['full_70_79'] + c['full_80']+c['full_missing']
```

```python
c['booster'] = c['booster_5_11'] + c['booster_12_17'] + c['booster_18_29'] + c['booster_30_39'] + c['booster_40_49'] + c['booster_50_59'] + c['booster_60_69'] + c['booster_70_79'] + c['booster_80']+c['booster_missing']
```

```python
c['booster2'] = c['booster2_5_11'] + c['booster2_12_17'] + c['booster2_18_29'] + c['booster2_30_39'] + c['booster2_40_49'] + c['booster2_50_59'] + c['booster2_60_69'] + c['booster2_70_79'] + c['booster2_80']++c['booster2_missing']
```

```python

c = c.reset_index()
```

```python
c = c.drop(columns=['partial_5_11','full_5_11','booster_5_11','booster2_5_11','partial_12_17','full_12_17','booster_12_17','booster2_12_17','partial_18_29','full_18_29','booster_18_29','booster2_18_29','partial_30_39','full_30_39','booster_30_39','booster2_30_39','partial_40_49','full_40_49','booster_40_49','booster2_40_49','partial_50_59','full_50_59','booster_50_59','booster2_50_59','partial_60_69','full_60_69','booster_60_69','booster2_60_69','partial_70_79','full_70_79','booster_70_79','booster2_70_79','partial_80','full_80','booster_80','booster2_80','partial_missing','full_missing','booster_missing','booster2_missing'])
```

```python
c.to_csv("half_cleaned.csv", index=False)
```
